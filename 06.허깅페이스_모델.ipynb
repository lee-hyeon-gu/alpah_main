```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions

from bs4 import BeautifulSoup
import re

import sys
if 'ipykernel' in sys.modules:
    from tqdm.notebook import tqdm
else:
    from tqdm import tqdm

import pandas as pd
import time

from selenium import webdriver

options = webdriver.ChromeOptions()
#options.add_argument("--headless")
options.add_argument('--disable-dev-shm-usage')
options.add_argument("--no-sandbox")  # sandbox 주요 요소 처리불가-> nosandbox 처리 가능

headers={'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36 Edg/120.0.0.0'}
options.add_argument('headers')
driver = webdriver.Chrome(options=options)
```

```python
url='https://huggingface.co/models?pipeline_tag=unconditional-image-generation&sort=likes'
driver.get(url)

all_title=[]
all_news=[]
all_url=[]
all_days=[]
all_images=[]
all_images_url=[]
all_videos=[]
all_videos_url=[]

titles=driver.find_elements(By.TAG_NAME,'h4')
day=driver.find_elements(By.TAG_NAME,'time')

for i in day:
    all_days.append(i.text) 
#    all_days.append(i.text.replace(', 2023','').replace(', 2022','')) 

    
for i in titles:
    all_title.append(i.text)

count=0# 횟수제한
for n in all_title:
    count+=1

    print(n)
    driver.find_element(By.PARTIAL_LINK_TEXT,n).click() #페이지 들어가기
    time.sleep(1)

    news=driver.find_elements(By.XPATH,'/html/body/div/main/div[2]/section[1]/div[3]/div[2]')
    for i in news:
        all_news.append(i.text)

    nourl=driver.current_url#url 저장
    all_url.append(nourl)
    time.sleep(1)

    if driver.find_elements(By.TAG_NAME,'img'): #이미지
        all_images.append('yes')
        image=driver.find_element(By.XPATH,'//img')
        img_srtc=image.get_attribute('src')
        all_images_url.append(img_srtc)
    else:
        all_images.append('no')
        all_images_url.append('-')


    # if driver.find_elements(By.TAG_NAME,'video'): # 동영상 model에는 동영상 없음
    #     all_images.append('yes')
    #     image=driver.find_element(By.XPATH,'동영상 xpath')
    #     img_srtc=image.get_attribute('src')
    #     all_images_url.append(img_srtc)
    # else:
    #     all_videos.append('no')
    #     all_videos_url.append('-')


    driver.back()
    time.sleep(1)
    if count>5:
        break

sumnews=[]
for t,n,u,d,i,iu in zip(all_title,all_news,all_url,all_days,all_images,all_images_url):
    singles={'title':t,'news':n,'url':u,'day':d,'image':i,'image_url':iu}
    sumnews.append(singles)

import pandas as pd
news_df=pd.DataFrame(sumnews)
print(news_df)

import os
if not os.path.exists('models.csv'):
    news_df.to_csv('models.csv', index=False, mode='w', encoding='utf-8-sig')
else:
    news_df.to_csv('models.csv', index=False, mode='a', encoding='utf-8-sig', header=False)
```
